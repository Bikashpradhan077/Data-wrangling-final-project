# 3. Low Level Data Undertanding

```python
#     a. Find count of unique values in location column.
#     b. Find which continent has maximum frequency using values counts.
#     c. Find maximum & mean value in 'total_cases'.
#     d. Find 25%,50% & 75% quartile value in 'total_deaths'.
#     e. Find which continent has maximum 'human_development_index'.
#     f. Find which continent has minimum 'gdp_per_capita'.
```

```python
unique_val = len(pd.unique(users['location'])) 
conti_maxfreq = users['continent'].value_counts().idxmax()
max_totalcases = users['total_cases'].max()
mean_totalcases = users['total_cases'].mean()
quartile_totaldeaths = users['total_cases'].quantile([0.25,0.50,0.75])
conti_maxindex = users.groupby('continent')['human_development_index'].max().idxmax()
conti_mingdp = users.groupby('continent')['gdp_per_capita'].min().idxmin()

print("Number of unique values in location column:", unique_val)
print("-" * 120)
print("Continent which has maximum frequency:", conti_maxfreq)
print("-" * 120)
print("Maximum Value in 'total_cases':", max_totalcases)
print("-" * 120)
print("Mean Value in 'total_cases':", mean_totalcases)
print("-" * 120)
print("25%, 50% and 75% quartile value in 'total_deaths':", quartile_totaldeaths)
print("-" * 120)
print("Continent which has maximum 'human_development_index':", conti_maxindex)
print("-" * 120)
print("Continent which has minimum 'gdp_per_capita':", conti_mingdp)
```

# 4. Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
data = users[['continent','location','date','total_cases','total_deaths',"gdp_per_capita"]]
data
```

# 5. Data Cleaning

```python
#     a. Remove all duplicates observations
#     b. Find missing values in all columns
#     c. Remove all observations where continent column value is missing
#         Tip : using subset parameter in dropna
#     d. Fill all missing values with 0
```

```python
users[['continent','location','date','total_cases','total_deaths',"gdp_per_capita",'human_development_index']].drop_duplicates(['continent','location'])
data
```

```python
data.isnull().sum()
```

```python
data.dropna(axis = 0, how = 'any', subset = 'continent', inplace = False)
```

```python
data = data.fillna(0)
data
```

# 6. Date time format

```python
#     a. Convert date column in datetime format using pandas.to_datetime
#     b. Create new column month after extracting month data from date column.
```

```python
print(data.isna().sum())
data = data.fillna(0)
data
```

# 7. Data Aggregation

```python
#     a. Find max value in all columns using groupby function on 'continent' column
#         Tip: use reset_index() after applying groupby
#     b. Store the result in a new dataframe named 'df_groupby'.
#         Use df_groupby dataframe for all further analysis
```

```python
maxval_col = users.groupby('continent').max().reset_index()
maxval_col
```

# 8. Feature Engineering

```python
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
```

```python
data["total_deaths_to_total_cases"] = data['total_cases']/data['total_deaths']
data["total_deaths_to_total_cases"].sample(10)
```

# 9. Data Visualization

```python
#     a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
#     b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
#     c. Plot Pairplot on df_groupby dataset.
#     d. Plot a bar plot of 'continent' column with 'total_cases' .
#         Tip : using kind='bar' in seaborn catplot
```

```python
sns.distplot(data['gdp_per_capita'],bins = 20 , kde = True )
plt.xlabel('gdp_per_capita')
plt.title('Histrogram and Density plot of GDP per Capita')
plt.show()
```

```python
plt.scatter(data["gdp_per_capita"], data['total_cases'])
plt.xlabel('Total Cases')
plt.ylabel('GDP Per Capita')
plt.title('scatter plot of total_cases & gdp_per_capita'.upper())
```
